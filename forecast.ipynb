```python
from fastapi import FastAPI, Form
from pydantic import BaseModel
import joblib
import pandas as pd

app = FastAPI()

# Load the saved model
saved_model = joblib.load('best_model.pkl')

# Define the request body model
class WeatherData(BaseModel):
    maxtempC: float
    mintempC: float
    cloudcover: float
    humidity: float
    sunHour: float
    HeatIndexC: float
    precipMM: float
    pressure: float
    windspeedKmph: float

# Define the route for prediction
@app.post("/predict")
async def predict(weather_data: WeatherData):
    # Convert request body to DataFrame
    input_data = pd.DataFrame(weather_data.dict(), index=[0])

    # Make prediction
    prediction = saved_model.predict(input_data)

    # Return prediction
    return {"prediction": prediction[0]}

# Define the root route
@app.get("/")
async def read_root():
    return {"message": "Welcome to Temperature Prediction API!"}
```
